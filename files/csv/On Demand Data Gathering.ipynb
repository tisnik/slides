### Loading data files

```python
# querying one record only (with insert working continuously)
times_one_record=pd.read_csv("query_times_one_record.csv", header=None)

# querying all records for a cluster (with insert working continuously)
times_all_records=pd.read_csv("query_times_all_records.csv", header=None)

# querying all records for a cluster (w/o inserts)
times_no_inserts=pd.read_csv("query_times_no_inserts.csv", header=None)
```
